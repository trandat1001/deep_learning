```python
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im
classifier = load_model('model/question1.h5')
test_img_path = 'image/prediction/banana_pineapple_strawberry_2.jpeg'
test_img = image.load_img(test_img_path, target_size=(64, 64, 3))
test_img = image.img_to_array(test_img)
test_img.shape
test_img = np.expand_dims(test_img, axis=0)
```

```python
result = classifier.predict_classes(test_img)
```

```python
# {'banana': 0, 'pineapple': 1, 'strawberry': 2}
if (result[0] == 0):
    print("Banana")
elif (result[0] == 1):
    print("Pineapple")
else:
    print("Strawberry")
```
