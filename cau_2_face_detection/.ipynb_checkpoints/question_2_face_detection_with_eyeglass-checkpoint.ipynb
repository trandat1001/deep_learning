```python
#import OpenCV library
import cv2
#import matplotlib library
import matplotlib.pyplot as plt
#importing time library for speed comparisons of both classifiers
import time 
import os
import math
%matplotlib inline
```

## We use openCV to build model to delect face straight with both glasses and no-glasses

```python
haar_face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_default.xml')
eye_glass_cascade = cv2.CascadeClassifier('data/haarcascade_eye_tree_eyeglasses.xml')
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def detect_faces(f_cascade, e_cascade, colored_img, scaleFactor = 1.11):
    #just making a copy of image passed, so that passed image is not changedÂ 
    img_copy = colored_img.copy()

    #convert the test image to gray image as opencv face detector expects gray images
    gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY) 

    #let's detect multiscale (some images may be closer to camera than others) images
    faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=5)   
    faces_found = len(faces)

    #go over list of faces and draw them as rectangles on original colored img
    for (x, y, w, h) in faces:
        cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 4)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img_copy[y:y+h, x:x+w]
        eyes = e_cascade.detectMultiScale(roi_gray,scaleFactor=scaleFactor, minNeighbors=3)
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 0, 255), 4)

    return img_copy, faces_found
```

```python
# Get all image files from 'img' folder and apply face and eye glass detection
image_folder = "data/img/"
files = os.listdir(image_folder)
fig = plt.figure(figsize=(128, 128))
columns = 2
rows = 4
for i in range(len(files)):
    img = cv2.imread(image_folder + files[i])
    faces_detected_img, faces_found = detect_faces(haar_face_cascade, eye_glass_cascade, img)
    fig.add_subplot(rows, columns, i + 1)
    plt.imshow(convertToRGB(faces_detected_img))
```
